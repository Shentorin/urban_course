```python
gdf.explore()
```

```python
DISTRICT = ox.geocode_to_gdf('Невский район, Санкт-Петербург')
okn_object = gdf.loc[0, "geometry"]

mask = gdf.geometry.within(DISTRICT.loc[0, "geometry"])
mask
```

```python
mask.value_counts()
```

```python
m = district.explore(style_kwds={"fill": False})
m = gdf[mask].explore(m=m, color="green")
m = gdf[~mask].explore(m=m, color="red")

m
```

```python
ENTERPRISE_BUFFER_SIZE = {4: 150}

enterprise_buffers = gdf.to_crs(3857).buffer(gdf["Category"].replace(ENTERPRISE_BUFFER_SIZE)).to_crs(4326)
enterprise_buffers.explore()
```

```python
union_enterprise_buffers = enterprise_buffers.unary_union
district.geometry.difference(union_enterprise_buffers).explore()
```
